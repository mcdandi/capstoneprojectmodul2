## CEK DUPLIKAT

### Cek duplikat (apakah 1 koridor yang sama memiliki corridorID yang berbeda)

```python
duplicate_corridors = df.groupby('corridorName')['corridorID'].nunique()

same_corridors_id = duplicate_corridors[duplicate_corridors > 1]
same_corridors_id
```

Pemeriksaan Duplikat:
- Dengan duplicate_corridors, kita menghitung jumlah ID koridor unik untuk setiap nama koridor (corridorName).
- same_corridors_id menampilkan hanya nama-nama koridor yang memiliki lebih dari satu ID berbeda, menunjukkan bahwa terdapat potensi duplikasi pada data koridor tersebut.

```python
result_df = pd.DataFrame({'corridorName': same_corridors_id.index,
                          'uniqueCorridorID': same_corridors_id.values})

corridor_ids = []
for corridor_name in result_df['corridorName']:
    unique_ids = df[df['corridorName'] == corridor_name]['corridorID'].unique()
    corridor_ids.append(unique_ids)
result_df['corridorID'] = corridor_ids
result_df
```

corridorID Unique Mapping:
- result_df menampilkan tabel dengan corridorName, jumlah ID unik (uniqueCorridorID), dan daftar ID koridor unik (corridorID).
- Ini menunjukkan bahwa beberapa nama koridor (misalnya, "Blok M - Kota") memiliki ID berbeda, seperti [M1, 1] untuk satu nama koridor yang sama.

```python
df.loc[(df['corridorName'] == 'Blok M - Kota') & (df['corridorID'] == 'M1'), 'corridorName'] = 'Blok M - Kota(M)'
df.loc[(df['corridorName'] == 'Kalideres - Bundaran HI via Veteran') & (df['corridorID'] == 'M3'), 'corridorName'] = 'Kalideres - Bundaran HI via Veteran(M)'
df.loc[(df['corridorName'] == 'Matraman Baru - Ancol') & (df['corridorID'] == 'M5'), 'corridorName'] = 'Matraman Baru - Ancol(M)'
df.loc[(df['corridorName'] == 'Pinang Ranti - Pluit') & (df['corridorID'] == 'M9'), 'corridorName'] = 'Pinang Ranti - Pluit(M)'
df.loc[(df['corridorName'] == 'Pulo Gadung - Monas') & (df['corridorID'] == 'M2'), 'corridorName'] = 'Pulo Gadung - Monas(M)'
```

Setelah melakukan pengecekan duplikasi, ditemukan bahwa terdapat beberapa koridor dengan nama yang sama tetapi memiliki corridorID unik yang berbeda. Perbedaan ini terjadi karena koridor tersebut dibedakan berdasarkan jam operasionalnya, misalnya ada koridor dengan jadwal operasional malam hari yang menggunakan kode khusus. Untuk menghindari kebingungan, koridor yang memiliki jam operasional malam diberi tambahan kode "(M)" di akhir nama koridor, misalnya Blok M - Kota menjadi Blok M - Kota(M). Hal ini membantu memastikan pembacaan data yang lebih jelas dan akurat sesuai dengan jadwal operasional masing-masing koridor.
